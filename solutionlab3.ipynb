```python
#Show a plot of the response rate by the sales channel.

filter = (data['Sales Channel'] == 'Agent') & (data['Response']=='Yes')
agentyes = len(data[filter])
print('Agent with Yes: ', agentyes)

filter = (data['Sales Channel'] == 'Agent') & (data['Response']=='Yes') | (data['Response']=='No')
agenttotal = len(data[filter])
print('Agent Total:', agenttotal)

filter = (data['Sales Channel'] == 'Branch') & (data['Response']=='Yes')
branch = len(data[filter])
print('Branch with Yes', branch)

filter = (data['Sales Channel'] == 'Branch') & (data['Response']=='Yes')  | (data['Response']=='No')
branchtotal = len(data[filter])
print('Branch Total', branchtotal)

filter = (data['Sales Channel'] =='Call Center')& (data['Response']=='Yes')
callc = len(data[filter])
print('Call Center with Yes', callc)

filter = (data['Sales Channel'] =='Call Center')& (data['Response']=='Yes') | (data['Response']=='No')
callcentertotal = len(data[filter])
print('Call Center Total', callcentertotal)

filter = (data['Sales Channel'] =='Web')& (data['Response']=='Yes')
web = len(data[filter])
print('Web with Yes', web)

filter = (data['Sales Channel'] =='Web')& (data['Response']=='Yes') | (data['Response']=='No')
webtotal = len(data[filter])
print('Web Total', webtotal)
#Agent          3477
#Branch         2567
#Call Center    1765
#Web            1325
```

```python
fig, ax = plt.subplots()

channels = ['Agent', 'Branch', 'Call Center', 'Web']
counts = [agentyes/agenttotal, branch/branchtotal, callc/callcentertotal, web/webtotal]
bar_labels = ['red', 'blue', 'green', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:green', 'tab:orange']

ax.bar(channels, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('Response rate')

ax.legend(title='Response rate by the sales channel')

plt.show()
```
